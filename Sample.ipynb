```python
import matplotlib.pyplot as plt

# unused but required import for doing 3d projections with matplotlib < 3.2
import mpl_toolkits.mplot3d  # noqa: F401

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
import pandas as pd
import numpy as np
import glob
pd.set_option('max_rows', 100)
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

df["target"] = iris.target
```

```python
df
```

```python
df['target'].value_counts()
```

```python
df.loc[(df.target=='0') ,'target'] = '0'
```

```python
df.loc[~(df.target=='0') ,'target'] = '1'
```

```python
df
```

```python
df['target'].value_counts()
```

```python
df['target'] = df['target'].astype(int)
df['sepal length (cm)'] = df['sepal length (cm)'].astype(int)
df['sepal width (cm)'] = df['sepal width (cm)'].astype(int)
df['petal length (cm)'] = df['petal length (cm)'].astype(int)
df['sepal width (cm)'] = df['sepal width (cm)'].astype(int)
```

# split data

```python
X=df.iloc[:, :-1]
y=df.target
```

```python
from sklearn.model_selection import train_test_split

# Without stratification divide into first partition and test set.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

# Hyper parameter

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
```

# SVC

```python
from sklearn.model_selection import GridSearchCV

    
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
 
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
 
# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
```

# Decision Tree

```python
param_grid1 = {'criterion': ['gini', 'entropy'],
              'max_depth': [2,3,4]}
 
grid1 = GridSearchCV(DecisionTreeClassifier(), param_grid1, refit = True, verbose = 3)
 
# fitting the model for grid search
grid1.fit(X_train, y_train)
```

```python
print(grid1.best_params_)
```

# Logistic Regression

```python
LogisticRegression().get_params().keys()
```

```python
param_grid2 = {'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
              'penalty': ['l1', 'l2'],
              'C': [0.1, 1, 10, 100, 1000]}
grid2 = GridSearchCV(LogisticRegression(), param_grid2, refit = True, verbose = 3)
 
# fitting the model for grid search
grid2.fit(X_train, y_train)
```

```python
print(grid2.best_params_)
```

# Random Forest

```python
param_grid3 = {'bootstrap': [True, False],
 'max_depth': [10,30,50],
 'max_features': ['auto', 'sqrt'],
 'min_samples_leaf': [1, 2, 4],
 'min_samples_split': [2, 5, 10],
 'n_estimators': [200, 400, 600]}


grid3 = GridSearchCV(RandomForestClassifier(), param_grid3, refit = True, verbose = 3)
 
# fitting the model for grid search
grid3.fit(X_train, y_train)
```

```python
print(grid3.best_params_)
```

# K-NN

```python
param_grid4 = {'n_neighbors': (1,10, 20),
    'leaf_size': (20,40,1),
    'p': (1,2),
    'weights': ('uniform', 'distance'),
    'metric': ('minkowski', 'chebyshev')}


grid4 = GridSearchCV(KNeighborsClassifier(), param_grid4, refit = True, verbose = 3)
 
# fitting the model for grid search
grid4.fit(X_train, y_train)
```

```python
print(grid4.best_params_)
```

# GaussianNB

```python
param_grid5 ={'var_smoothing': np.logspace(0,-9, num=100)}


grid5 = GridSearchCV(GaussianNB(), param_grid5, refit = True, verbose = 3)
 
# fitting the model for grid search
grid5.fit(X_train, y_train)
```

```python
print(grid5.best_params_)
```

# Gradient Boosting

```python
param_grid6 = {"n_estimators":[5,50,250,500],
    "max_depth":[1,3,5,7,9],
    "learning_rate":[0.01,0.1,1,10,100]}


grid6 = GridSearchCV(GradientBoostingClassifier(), param_grid6, refit = True, verbose = 3)
 
# fitting the model for grid search
grid6.fit(X_train, y_train)
```

```python
print(grid6.best_params_)
```

# Model

```python
clf1 = SVC(C= 0.1, gamma=1, kernel='rbf')
clf2 = RandomForestClassifier(bootstrap=True, max_depth= 10,max_features='auto', min_samples_leaf= 1, min_samples_split= 2, n_estimators= 200)
clf3 = KNeighborsClassifier(leaf_size= 20, metric='minkowski', n_neighbors=1, p= 1, weights='uniform')
clf4 = LogisticRegression(C=0.1, penalty='l1', solver='liblinear')
clf5 = GaussianNB(var_smoothing= 1)
ensemble = GradientBoostingClassifier(n_estimators=50, learning_rate=0.01,max_depth=1)
```

```python
predictions1 = clf1.predict(X_test)
```
