```python
df = pd.read_csv('C:/Users/huawei/Desktop/brs.csv',header=None,names=['xs','ys','yc'])
```

```python
X=df.iloc[:, :-1]
y=df.yc
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV

    
# defining parameter range
param_grid = {'C': [0.5, 1, 2, 4, 8],
              'gamma': [1, 0.5, 0.25, 0.125, 0.0625],
              'kernel': ['rbf']}
 
grid = GridSearchCV(SVC(), param_grid,cv=100, refit = True, verbose = 3)
 
# fitting the model for grid search
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
```

```python
# defining parameter range
param_grid = {'C': [8,16,32,64,128],
              'gamma': [0.5,1,2,4,8 ],
              'kernel': ['rbf']}
 
grid = GridSearchCV(SVC(), param_grid,refit='accuracy',scoring = ['f1','accuracy','recall','precision'],cv=100, verbose = 3)
 
# fitting the model for grid search
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
```

```python
  print("\n The best score(accuracy) across ALL searched params:\n",grid.best_score_)
```

```python
print('Test Accuracy on test data: ', grid.score(X_test,y_test))
```

```python
model2 = SVC(C=16, gamma=2, kernel='rbf')
```

```python
accuracy = cross_val_score(model2, X,y, scoring='accuracy', cv = 10)
f1=cross_val_score(model2, X,y, scoring='f1', cv = 10)
pre=cross_val_score(model2, X,y, scoring='precision', cv = 10)
recall=cross_val_score(model2,X,y, scoring='recall', cv = 10)
print(accuracy)
#get the mean of each fold 
print("Accuracy of Model with Cross Validation is:",accuracy.mean() * 100)
print("F1 of Model with Cross Validation is:",f1.mean() * 100)
print("Precision of Model with Cross Validation is:",pre.mean() * 100)
print("Recall of Model with Cross Validation is:",recall.mean() * 100)
```
